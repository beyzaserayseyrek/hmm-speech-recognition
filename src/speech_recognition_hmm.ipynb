Bu çalışmada Hidden Markov Model (HMM) kullanılarak
basit bir kelime tanıma sistemi simüle edilmiştir.
İki kelime için model oluşturulmuş ve yeni gelen
gözlem dizisinin hangi kelimeye ait olduğu
olasılık hesapları ile belirlenmiştir.

```
import numpy as np
from hmmlearn import hmm
```

```
# 'EV' kelimesi icin
model_ev = hmm.MultinomialHMM(n_components=2)

model_ev.n_trials = 1

model_ev.startprob_ = np.array([1.0, 0.0])

model_ev.transmat_ = np.array([ [0.6, 0.4],
                                [0.2, 0.8] ])

model_ev.emissionprob_ = np.array([
    [0.7, 0.3],
    [0.1, 0.9]
])
```

```
# OKUL kelimesi icin
model_okul = hmm.MultinomialHMM(n_components=2)

model_okul.n_trials = 1

model_okul.startprob_ = np.array([
    1.0,
    0.0
])

model_okul.transmat_ = np.array([
    [0.5, 0.5],
    [0.4, 0.6]
])

model_okul.emissionprob_ = np.array([
    [0.6, 0.4],
    [0.3, 0.7]
])
```

```
# Observation encoding
# High -> 0 (becomes [1, 0] count vector)
# Low  -> 1 (becomes [0, 1] count vector)

test = np.array([[1, 0], [0, 1]])
```

```
score_ev = model_ev.score(test)
score_okul = model_okul.score(test)

print("EV model score:", score_ev)
print("OKUL model score:", score_okul)
```

```
if score_ev > score_okul:
    print("Tahmin edilen kelime: EV")
else:
    print("Tahmin edilen kelime: OKUL")
```

```
def predict_word(observation):
    score_ev = model_ev.score(observation)
    score_okul = model_okul.score(observation)

    print("EV model score:", score_ev)
    print("OKUL model score:", score_okul)

    if score_ev > score_okul:
        return "EV"
    else:
        return "OKUL"
```

```
prediction = predict_word(test)
print("Tahmin edilen kelime:", prediction)
```

```
# farklı gözlem örnekleri

test1 = np.array([[1,0],[0,1]])  # High Low
test2 = np.array([[1,0],[1,0]])  # High High
test3 = np.array([[0,1],[0,1]])  # Low Low

print("Test1:", predict_word(test1))
print("Test2:", predict_word(test2))
print("Test3:", predict_word(test3))
```
