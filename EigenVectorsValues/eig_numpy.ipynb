# NumPy eig Fonksiyonu Kullanımı

Bu çalışmada NumPy kütüphanesinin eig fonksiyonu kullanılarak bir matrisin özdeğerleri ve özvektörleri hesaplanmaktadır.


```python
import numpy as np

# örnek matris
A = np.array([[2, 1],
              [1, 2]])

print("Matrix A:")
print(A)
```

## NumPy eig Fonksiyonu

NumPy kütüphanesi, özdeğer ve özvektör hesaplamak için linalg.eig fonksiyonunu sağlar.

```python
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)
```

## Sonuç

Elde edilen özdeğerler ve özvektörler, matrisin belirli yönlerde farklı oranlarda ölçekleme yaptığını göstermektedir.

Özvektörler, matris uygulandığında yönü değişmeyen vektörleri temsil ederken, özdeğerler bu yönlerdeki ölçekleme miktarını ifade eder.
