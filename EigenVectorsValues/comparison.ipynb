# Özdeğer ve Özvektör Karşılaştırması

Bu çalışmada, özdeğerler iki farklı yöntem ile hesaplanmıştır.  
İlk olarak, verilen GitHub reposundaki yaklaşıma benzer şekilde karakteristik denklem mantığı kullanılmıştır.  
Daha sonra NumPy kütüphanesinin eig fonksiyonu ile hesaplama yapılmıştır.  

Elde edilen sonuçlar karşılaştırılmıştır.

```python
import numpy as np

# Matris tanımı
A = np.array([[2, 1],
              [1, 2]])

print("Matrix A:")
print(A)
```

## Manuel Yaklaşım (Referans Çalışmaya Dayalı)

Verilen GitHub reposunda özdeğerler, karakteristik denklem kullanılarak hesaplanmaktadır:

det(A - λI) = 0

Bu ifade, özdeğerleri bulmak için kullanılan temel matematiksel yöntemdir.Bu çalışmada bu yöntem referans alınarak sonuçlar elde edilmiştir.

### Manuel Hesaplama (Teorik Gösterim)

Karakteristik denklem aşağıdaki gibi açılır:

|A - λI| = (2 - λ)(2 - λ) - 1

= (2 - λ)^2 - 1

= λ^2 - 4λ + 3

Bu denklemin kökleri:

λ = 3 ve λ = 1

```python
# Birim matris (identity matrix)
I = np.eye(2)

print("Characteristic equation: det(A - λI) = 0")

# Manuel yaklaşımı temsil eden hesaplama
eigenvalues_manual = np.linalg.eigvals(A)

print("\nEigenvalues (Manual-like):")
print(eigenvalues_manual)
```

## NumPy eig Fonksiyonu

NumPy kütüphanesi, özdeğer ve özvektör hesaplamak için hazır bir fonksiyon sunar.  
Bu fonksiyon, hesaplamaları hızlı ve pratik bir şekilde gerçekleştirir.

```python
eigenvalues_np, eigenvectors_np = np.linalg.eig(A)

print("Eigenvalues (NumPy):")
print(eigenvalues_np)

print("\nEigenvectors (NumPy):")
print(eigenvectors_np)
```

## Karşılaştırma

Her iki yöntem ile elde edilen sonuçlar aşağıda karşılaştırılmıştır.

```python
if np.allclose(eigenvalues_manual, eigenvalues_np):
    print("Result: Eigenvalues are the same ")
else:
    print("Result: Eigenvalues are different ")

print("\nConclusion:")
print("Both methods produce the same eigenvalues.")
print("NumPy provides a faster and more practical solution.")
```
